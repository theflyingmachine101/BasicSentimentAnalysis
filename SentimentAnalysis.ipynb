```python
import pandas as pd
TrainingData=pd.read_csv('test.csv')
TrainingData=TrainingData[:1000]
TrainingData.shape
```

```python
TrainingData.head()
X=TrainingData.iloc[:,1:2]
X=X.values.tolist()
X
```

```python
import textblob as tb
polarity=0.0

for i in range(1000):
    w=tb.TextBlob(X[i][0])
    polarity+=w.sentiment.polarity
#    print(w.sentiment.polarity)
```

```python
polarity=polarity/10
print(polarity)
#indicates the given tweets  are moderately positive
```
